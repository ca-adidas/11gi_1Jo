```python
print(train.columns)
```

id: 각 버섯 샘플에 대한 고유 식별자입니다.

class: 목표 변수로, 버섯이 독성이 있는지(예: 'poisonous') 또는 먹을 수 있는지(예: 'edible')를 나타냅니다. 이진 분류에서는 '0'은 먹을 수 있는 경우, '1'은 독성인 경우를 의미할 수 있습니다.

cap-diameter: 버섯 갓의 지름을 나타냅니다.

cap-shape: 버섯 갓의 모양을 나타냅니다 (예: 종 모양, 납작한 모양 등).

cap-surface: 버섯 갓의 표면 질감을 나타냅니다 (예: 매끄럽거나, 섬유질이 있거나, 비늘이 있는 등).

cap-color: 버섯 갓의 색상을 나타냅니다.

does-bruise-or-bleed: 버섯이 멍이 들거나 액체를 분비하는지 여부를 나타냅니다.

gill-attachment: 버섯의 주름(gill)이 줄기에 어떻게 붙어있는지 나타냅니다 (예: 부착됨, 떨어져 있음 등).

gill-spacing: 버섯 주름의 간격을 나타냅니다 (예: 촘촘함, 넓음).

gill-color: 버섯 주름의 색상을 나타냅니다.

stem-height: 버섯 줄기의 높이를 나타냅니다.

stem-width: 버섯 줄기의 너비를 나타냅니다.

stem-root: 버섯 줄기의 뿌리 부분에 대한 정보를 나타냅니다.

stem-surface: 버섯 줄기 표면의 질감을 나타냅니다 (예: 섬유질, 매끄러움 등).

stem-color: 버섯 줄기의 색상을 나타냅니다.

veil-type: 버섯을 덮고 있는 막(veil)의 유형을 나타냅니다.

veil-color: 버섯 막의 색상을 나타냅니다.

has-ring: 버섯 줄기에 고리가 있는지 여부를 나타냅니다.

ring-type: 버섯 줄기의 고리 유형을 나타냅니다.

spore-print-color: 버섯 포자의 색상을 나타냅니다.

habitat: 버섯이 자라는 서식지를 나타냅니다 (예: 숲, 목초지 등).

season: 버섯이 발견되는 계절을 나타냅니다.

```python
# 문자 컬럼별 unique한 값들
for column in train.columns:
  if train[column].dtype == object:
    print(f"Column: {column}")
    print(train[column].unique())
    print("\n")
```
